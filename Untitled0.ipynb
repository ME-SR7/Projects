```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
data={
    'Month':pd.date_range(start='2000-01-01', periods=36, freq='M'),
    'Sales':np.random.randint(200,530,36)
}

df=pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(12,10))
plt.plot(df['Sales'], marker='o')
plt.title("Sales Over Time")
plt.ylabel("Sale")
plt.xlabel("Months")
plt.grid(True)
plt.show()
```

#### Decompose the Time Series

##### To understand trend, seasonality, and residual parts separately:

```python
result=seasonal_decompose(df['Sales'], model='additive', period=12)
result.plot()
plt.show()
```
